## 1.2 

Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
delta = df['yr_renovated'] - df['yr_built'] # заводим переменную, в которой будем хранить разницу между годом реновацией дома и годом постройки дома.
df['delta_renovated'] = delta.apply(lambda x: x if x > 0 else 0) # создаем признак, который будет содержать разницу между годом реновацией дома и годом постройки дома.
```

```python
df['delta_renovated'].unique() # чтобы показать, что нет минусовых значений. 
```

```python
df[df['delta_renovated']>0] # чтобы показать, что получилось создать 'delta_renovated' с правильными значениями.
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_sale'] = df['date'].apply(lambda x: int(x[:4])) # признак года продажи (конвертировал значение в int)
df.head() 
```

```python
df['month_sale'] = df['date'].apply(lambda x: int(x[4:6])) # признак месяца продажи (конвертировал значение в int)
df.head() 
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True) # удалил признаки date, zipcode, lat, long
df.head()
```

## Задание 3 Составьте несколько сводных таблиц

## 3.1 

Найдите среднюю стоимость домов в зависимости от количества спален

Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

## 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['mean', 'min', 'max']}) # mean - среднее, min - минимальное, max - максимальное, condition - состояние дома. 
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index='waterfront', columns='view',
               aggfunc='count', fill_value=0,
               values='id', margins=True) # waterfront - вид на набережную, view - оценка вида.
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=df['floors'], columns=df['bedrooms']) # больше всего домов с 3 спальнями с 1 этажом = 5455 домов. 
```

### Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.pivot_table(index='condition', columns='grade',
               aggfunc='median', fill_value=0,
               values='price', margins=True) # condition - состояние дома, grade - оценка дома.
```
